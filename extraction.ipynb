```python
import time
import logging
import concurrent.futures


# Create an empty list to hold movie data
movies_data = []

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

def fetch_movie_data(movie_id):
    try:
        movie = ia.get_movie(movie_id)
        ia.update(movie, 'reviews')  # Fetch reviews for the movie
        reviews = movie.get('reviews', [])
        return {
            'movie_id': movie_id,
            'reviews': [{'content': review['content'], 'rating': review.get('rating')} for review in reviews]
        }
    except Exception as e:
        logging.error(f"Error fetching movie ID {movie_id}: {e}")
        return None

# Define a function to process a batch of movie IDs
def process_batch(batch_ids, batch_num, total_batches):
    batch_data = []
    logging.info(f"Fetching batch {batch_num}/{total_batches} with {len(batch_ids)} movies")
    for movie_id in batch_ids:
        movie_data = fetch_movie_data(movie_id)
        if movie_data:
            batch_data.append(movie_data)
    return batch_data

# Number of threads to use
num_threads = 8

# Batch size
batch_size = 100

# Split the list of movie IDs into batches
batches = [movie_ids[i:i + batch_size] for i in range(0, len(movie_ids), batch_size)]
num_batches = len(batches)

# Process batches concurrently
with concurrent.futures.ThreadPoolExecutor(max_workers=num_threads) as executor:
    future_to_batch = {executor.submit(process_batch, batch, i + 1, len(batches)): batch for i, batch in enumerate(batches)}
    for future in concurrent.futures.as_completed(future_to_batch):
        batch_data = future.result()
        if batch_data:
            movies_data.extend(batch_data)
        # Optional: Add delay to avoid rate limiting
        time.sleep(1)

# Convert the list of dictionaries to a DataFrame
movies_df = pd.DataFrame(movies_data)
```
