```python
import pandas as pd
```

```python
# Load the datasets

reviews_df = pd.read_csv('imdb_reviews_data.csv')
hydra_df = pd.read_csv('Hydra-Movie-Scrape.csv')
```

```python
print(reviews_df.head())
print(hydra_df.head())
```

```python
# Remove the leading 'tt' from the IMDB ID
hydra_df['IMDB ID'] = hydra_df['IMDB ID'].str[2:]
hydra_df.head()
```

```python
genres_df = hydra_df[['IMDB ID', 'Genres']].copy()
genres_df = genres_df.assign(Genres=genres_df['Genres'].str.split('|')).explode('Genres').reset_index(drop=True)
genres_df.head()
```

```python
cast_df = hydra_df[['IMDB ID', 'Cast']].copy()
cast_df = cast_df.assign(Cast=cast_df['Cast'].str.split('|')).explode('Cast').reset_index(drop=True)
cast_df.head()
```

```python
movies_df = hydra_df.drop(columns=['Genres', 'Cast'])
movies_df.head()
```

```python
# Example query: Find all movies in the "Comedy" genre
comedy_movies = genres_df[genres_df['Genres'] == 'Comedy']
comedy_movies_with_details = comedy_movies.merge(movies_df, on='IMDB ID')

print("Comedy Movies with Details:\n", comedy_movies_with_details)
```
